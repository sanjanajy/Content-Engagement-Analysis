```python
# Install the required library
!pip install ucimlrepo

# Import necessary libraries
from ucimlrepo import fetch_ucirepo
import pandas as pd

# Fetch the dataset
online_news_popularity = fetch_ucirepo(id=332)

# Extract features (X) and target (y)
X = online_news_popularity.data.features
y = online_news_popularity.data.targets

# Convert to Pandas DataFrame
df = pd.concat([X, y], axis=1)

# Display basic information
print(df.info())
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
print(df.columns)

```

```python
# Remove leading and trailing spaces in column names
df.columns = df.columns.str.strip()

# Now drop unnecessary columns
df_cleaned = df.drop(columns=['LDA_00', 'LDA_01', 'LDA_02', 'LDA_03', 'LDA_04',
                              'self_reference_min_shares', 'self_reference_max_shares',
                              'is_weekend'])

# Display updated column names
print(df_cleaned.columns)
```

```python
# Rename columns for better understanding
df_cleaned = df_cleaned.rename(columns={
    'n_tokens_title': 'title_word_count',
    'n_tokens_content': 'content_word_count',
    'n_unique_tokens': 'unique_word_ratio',
    'n_non_stop_words': 'non_stop_word_ratio',
    'n_non_stop_unique_tokens': 'unique_non_stop_word_ratio',
    'num_hrefs': 'num_links',
    'num_self_hrefs': 'num_internal_links',
    'num_imgs': 'num_images',
    'num_videos': 'num_videos',
    'average_token_length': 'avg_word_length',
    'num_keywords': 'num_keywords',
    'global_subjectivity': 'subjectivity_score',
    'global_sentiment_polarity': 'sentiment_score',
    'shares': 'shares_count'  # Target variable
})

# Display the new column names
print(df_cleaned.columns)
```

```python
# Summary statistics of numerical features
print(df_cleaned.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a modern theme for better visuals
sns.set_style("whitegrid")

# Plot distribution of shares_count
plt.figure(figsize=(10, 5))
sns.histplot(df_cleaned['shares_count'], bins=50, kde=True, color="dodgerblue")

# Improve readability
plt.axvline(df_cleaned['shares_count'].median(), color='red', linestyle='dashed', label="Median")
plt.xlabel("Number of Shares")
plt.ylabel("Frequency")
plt.title("ðŸ“Š Distribution of Article Shares")
plt.legend()
plt.show()

```

```python
import numpy as np

# Apply log transformation (Adding 1 to avoid log(0))
df_cleaned['log_shares'] = np.log1p(df_cleaned['shares_count'])

# Plot the transformed distribution
plt.figure(figsize=(10, 5))
sns.histplot(df_cleaned['log_shares'], bins=50, kde=True, color="seagreen")

plt.axvline(df_cleaned['log_shares'].median(), color='red', linestyle='dashed', label="Median")
plt.xlabel("Log(1 + Number of Shares)")
plt.ylabel("Frequency")
plt.title("ðŸ“Š Distribution of Log-Transformed Article Shares")
plt.legend()
plt.show()
```

```python
corr_matrix = df_cleaned.corr()
corr_matrix['log_shares'].sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(12, 8))

# Create heatmap
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, fmt=".2f", linewidths=0.5)

# Title
plt.title("Correlation Matrix Heatmap", fontsize=14)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

channel_cols = ['data_channel_is_lifestyle', 'data_channel_is_entertainment',
                'data_channel_is_bus', 'data_channel_is_socmed',
                'data_channel_is_tech', 'data_channel_is_world']

avg_shares_per_channel = {col: df_cleaned[df_cleaned[col] == 1]['shares_count'].mean() for col in channel_cols}

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=list(avg_shares_per_channel.keys()), y=list(avg_shares_per_channel.values()), palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel("Content Category")
plt.ylabel("Average Shares")
plt.title("Average Shares by Content Type")
plt.show()

# Print result
most_popular_channel = max(avg_shares_per_channel, key=avg_shares_per_channel.get)
print(f"ðŸ”¹ The content category with the highest average shares is **{most_popular_channel.replace('data_channel_is_', '').capitalize()}** with {avg_shares_per_channel[most_popular_channel]:.2f} shares on average.")


```

```python
weekday_cols = ['weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday',
                'weekday_is_thursday', 'weekday_is_friday', 'weekday_is_saturday', 'weekday_is_sunday']

avg_shares_per_day = {col: df_cleaned[df_cleaned[col] == 1]['shares_count'].mean() for col in weekday_cols}

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=list(avg_shares_per_day.keys()), y=list(avg_shares_per_day.values()), palette='viridis')
plt.xticks(rotation=45)
plt.xlabel("Day of the Week")
plt.ylabel("Average Shares")
plt.title("Engagement Based on Day of the Week")
plt.show()

# Print result
best_day = max(avg_shares_per_day, key=avg_shares_per_day.get).replace('weekday_is_', '').capitalize()
print(f"ðŸ”¹ The best day for publishing is **{best_day}**, with an average of {avg_shares_per_day['weekday_is_' + best_day.lower()]:.2f} shares.")
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x=df_cleaned['num_images'], y=df_cleaned['shares_count'], alpha=0.5, label='Images')
sns.scatterplot(x=df_cleaned['num_videos'], y=df_cleaned['shares_count'], alpha=0.5, label='Videos')
plt.xlabel("Number of Media Elements")
plt.ylabel("Shares")
plt.title("Impact of Images & Videos on Shares")
plt.legend()
plt.show()

# Print correlation result
image_corr = df_cleaned['num_images'].corr(df_cleaned['shares_count'])
video_corr = df_cleaned['num_videos'].corr(df_cleaned['shares_count'])

print(f"ðŸ”¹ Correlation with shares: **Images: {image_corr:.2f}, Videos: {video_corr:.2f}**")
if image_corr > video_corr:
    print("ðŸ”¹ More images tend to result in higher shares.")
else:
    print("ðŸ”¹ More videos tend to result in higher shares.")

```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x=df_cleaned['kw_avg_avg'], y=df_cleaned['shares_count'], alpha=0.5, label='Avg Keyword Strength')
sns.scatterplot(x=df_cleaned['kw_max_avg'], y=df_cleaned['shares_count'], alpha=0.5, label='Max Keyword Strength')
sns.scatterplot(x=df_cleaned['kw_min_avg'], y=df_cleaned['shares_count'], alpha=0.5, label='Min Keyword Strength')
plt.xlabel("Keyword Strength")
plt.ylabel("Shares")
plt.title("Impact of Keyword Strength on Shares")
plt.legend()
plt.show()

# Print result
kw_avg_corr = df_cleaned['kw_avg_avg'].corr(df_cleaned['shares_count'])
kw_max_corr = df_cleaned['kw_max_avg'].corr(df_cleaned['shares_count'])
kw_min_corr = df_cleaned['kw_min_avg'].corr(df_cleaned['shares_count'])

print(f"ðŸ”¹ Correlation with shares: Avg Keywords: {kw_avg_corr:.2f}, Max Keywords: {kw_max_corr:.2f}, Min Keywords: {kw_min_corr:.2f}")
if kw_avg_corr > kw_max_corr and kw_avg_corr > kw_min_corr:
    print("ðŸ”¹ A balanced keyword strategy (average keyword strength) is best for engagement.")
elif kw_max_corr > kw_avg_corr and kw_max_corr > kw_min_corr:
    print("ðŸ”¹ Using strong, high-impact keywords in articles results in more shares.")
else:
    print("ðŸ”¹ Even low keyword strength articles can perform well, meaning other factors may matter more.")
```
